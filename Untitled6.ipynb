```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('laptopscrapy.csv')

# 1. Фильтрация

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['review_count'] = pd.to_numeric(df['review_count'], errors='coerce')

filtered_df = df[(df['price'] > 0) & (df['screen_size'].notnull()) & (df['RAM'].notnull())]

filtered_df['brand'] = filtered_df['brand'].fillna('Unknown')

print("Фильтрованный набор данных:")
print(filtered_df.info())
print(filtered_df.describe())
```

```python
# 2. Группировка

grouped_df = filtered_df.groupby(['brand', 'operating_system']).agg({
    'price': ['mean', 'median'],
    'review_count': ['mean', 'median']
}).reset_index()

print("\nАгрегированные данные по брендам и ОС:")
print(grouped_df)
```

```python
# 3. Визуализация

plt.figure(figsize=(10, 6))
sns.barplot(x='brand', y=('price', 'mean'), data=grouped_df, palette='viridis')
plt.title('Средняя цена по брендам')
plt.xlabel('Бренд')
plt.ylabel('Средняя цена')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(filtered_df['price'], kde=True, bins=20, color='blue')
plt.title('Распределение цен')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()

plt.figure(figsize=(8, 6))

if filtered_df['brand'].nunique() > 1:
    sns.scatterplot(x='review_count', y='price', data=filtered_df, hue='brand', palette='deep')
else:
    print("Недостаточно уникальных значений в 'brand' для использования hue.")
    sns.scatterplot(x='review_count', y='price', data=filtered_df)

plt.title('Цена vs. Количество отзывов')
plt.xlabel('Количество отзывов')
plt.ylabel('Цена')
plt.tight_layout()
plt.show()
```

```python
# 4. Очистка

df_cleaned = filtered_df.copy()
df_cleaned['price'] = df_cleaned['price'].fillna(df_cleaned['price'].median())
df_cleaned['review_count'] = df_cleaned['review_count'].fillna(df_cleaned['review_count'].median())

Q1 = df_cleaned['price'].quantile(0.25)
Q3 = df_cleaned['price'].quantile(0.75)
IQR = Q3 - Q1
df_cleaned = df_cleaned[(df_cleaned['price'] >= Q1 - 1.5 * IQR) & (df_cleaned['price'] <= Q3 + 1.5 * IQR)]

print("\nОчищенный набор данных:")
print(df_cleaned.info())
print(df_cleaned.describe())
```
